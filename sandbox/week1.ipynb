```python
%matplotlib inline
import os, glob, cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
 
DATA_DIR = r"/home/karthik/projects/circuitguard/dataset/PCBData"

if not os.path.exists(DATA_DIR):
    raise FileNotFoundError(f"Dataset directory not found: {DATA_DIR}")
```

```python
groups = [d for d in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, d))]
print("Total groups:", len(groups))
print("First few groups:", groups[:5])
```

```python
 # %%
group = os.path.join(DATA_DIR, groups[0])   # pick first group
print("Using group:", group)

# Collect template images (any subfolder)
template_files = []
for sub in os.listdir(group):
    sub_path = os.path.join(group, sub)
    if os.path.isdir(sub_path) and not sub.lower().endswith("not"):
        template_files.extend(glob.glob(os.path.join(sub_path, "*.jpg")))

# Collect test/defective images (folders ending with '_not')
test_files = []
for sub in os.listdir(group):
    sub_path = os.path.join(group, sub)
    if os.path.isdir(sub_path) and sub.lower().endswith("_not"):
        test_files.extend(glob.glob(os.path.join(sub_path, "*.jpg")))

# Debug prints
print("Templates found:", template_files)
print("Tests found:", test_files)

if not template_files:
    raise FileNotFoundError(f"No template images found in {group}")
if not test_files:
    raise FileNotFoundError(f"No test images found in {group}")

# Pick first template and test
template_path = template_files[0]
test_path     = test_files[0]

print("Template:", template_path)
print("Test:", test_path)
```

```python
 # %%
template = cv2.imread(template_path, 0)
test     = cv2.imread(test_path, 0)

# Debug: check if images loaded
if template is None:
    raise ValueError(f"Failed to load template image: {template_path}")
if test is None:
    raise ValueError(f"Failed to load test image: {test_path}")

print("Template shape:", template.shape)
print("Test shape:", test.shape)
```
